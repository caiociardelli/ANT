```python
# Import libraries
import time
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt

from scipy.ndimage import uniform_filter1d
from obspy.signal.filter import bandpass, highpass
from obspy.core import UTCDateTime
from obspy import read, read_inventory
from obspy.signal.detrend import spline
```

```python
# Auxiliary functions
def cosine_taper(sgn, taper_length):
  """
  Applies a cosine taper to the input signal.

  Parameters:
  sgn (array-like): Input signal.
  taper_length (int): Length of the taper.

  Returns:
  array-like: Signal with cosine taper applied.
  """

  # Iterate over the taper length
  for i in range(taper_length):

    # Calculate the weight for the current sample
    w = 0.5 * (1.0 - np.cos(np.pi * float(i) / taper_length))

    # Apply the weight to the current and mirrored samples
    sgn[i]  *= w
    sgn[-i] *= w

  # Return the tapered signal
  return sgn

def cosine_taper_fraction(sgn, taper_fraction=0.01):
  """
  Applies a cosine taper to the input signal.

  Parameters:
  sgn (array-like): Input signal.
  taper_fraction (float): Fraction of the signal to taper.
  Default is 0.01.

  Returns:
  array-like: Signal with cosine taper applied.
  """
  # Calculate the taper length
  taper_length = int(taper_fraction * len(sgn))

  # Apply the cosine taper
  for i in range(taper_length):
    # Calculate the weight for the current sample
    w = 0.5 * (1.0 - np.cos(np.pi * float(i) / taper_length))

    # Apply the weight to the current and mirrored samples
    sgn[i]  *= w
    sgn[-i] *= w

  return sgn

def centered(array, n):
  """
  Returns a centered subset of an array.

  Parameters:
  array (array-like): The input array.
  n (int): The length of the centered subset.

  Returns:
  array-like: The centered subset of the input array.
  """
  # Calculate the starting index and ending index of the centered subset.
  size = len(array)
  i1 = (size - n) // 2  # integer division to get the floor value
  i2 = i1 + n

  # Return the centered subset of the array.
  return array[i1:i2]

def invert_spectrum(spectrum, freqs, freqmin, freqmax):
  """
  Invert the spectrum within the given frequency range.

  Parameters:
  spectrum (array-like): Input spectrum.
  freqs (array-like): Frequency values corresponding to the spectrum.
  freqmin (float): Minimum frequency value.
  freqmax (float): Maximum frequency value.

  Returns:
  array-like: Inverted spectrum.
  """

  # Find the indices of the frequencies within the given range.
  ii = ((freqs > freqmin) & (freqs < freqmax)).nonzero()[0]

  # Create an array to store the inverted spectrum.
  spectrum_inv = np.zeros(spectrum.size, dtype=complex)

  # Invert the spectrum within the given frequency range.
  spectrum_inv[ii] = 1.0 / spectrum[ii]

  # Return the inverted spectrum.
  return spectrum_inv

def removeResponse(sgn, response, freqs, freqmin, freqmax, nfft, taper_length, phase_only=False):
  """
  Removes the instrument response from the input signal.

  Parameters:
  sgn (array-like): Input signal.
  response (array-like): Frequency response of the instrument.
  freqs (array-like): Corresponding frequency values of the instrument
  response.
  freqmin (float): Minimum frequency value.
  freqmax (float): Maximum frequency value.
  nfft (int): FFT size.
  taper_length (int): Length of the cosine taper.
  phase_only (bool): If True, only invert the phase of the instrument
  response.

  Returns:
  array-like: Signal with the instrument response removed.
  """
  # Get the length of the input signal
  n = len(sgn)

  # Create a copy of the signal as a float64 array
  sgn_copy = sgn.astype(np.float64)

  # Remove the mean and apply a cosine taper
  sgn_copy = sp.signal.detrend(sgn_copy)
  sgn_copy = cosine_taper(sgn_copy, taper_length)

  # Perform the FFT
  sgn_fft = np.fft.rfft(sgn_copy, n=nfft)

  # Invert the phase of the instrument response if requested
  if phase_only:
    response_inverse = np.exp(-1j * np.angle(response))
  else:
    response_inverse = invert_spectrum(response, freqs, freqmin, freqmax)

  # Calculate the frequencies of the prefilter
  f1 = freqmin
  f2 = 1.4 * freqmin
  f3 = 0.9 * freqmax
  f4 = freqmax

  # Create the prefilter
  prefilter = np.zeros_like(freqs)

  prefilter[(freqs >= f1) & (freqs <  f2)] = (freqs[(freqs >= f1) & (freqs < f2)] - f1) / (f2 - f1)
  prefilter[(freqs >= f2) & (freqs <= f3)] = 1.0
  prefilter[(freqs >  f3) & (freqs <= f4)] = (f4 - freqs[(freqs > f3) & (freqs <= f4)]) / (f4 - f3)

  # Multiply the FFT of the signal by the inverse of the instrument response
  sgn_fft_response = prefilter * sgn_fft * response_inverse

  # Perform the inverse FFT and extract the real part
  sgn_response = np.fft.irfft(sgn_fft_response, n=nfft)[:n].real

  return sgn_response

def normalize(sgn, mode='no_normalization', clip_factor=6, clip_weight=10, norm_window=None,
              freqmin=0.01, sampling_rate=1.0, water_level=1e-100, taper_length=None):
  """
  Normalizes the input signal.

  Parameters:
  sgn (array-like): Input signal.
  mode (str): Normalization mode. Options are 'clipping',
              'clipping_iteration', 'running_average',
              '1bit', and 'no_normalization'.
  clip_factor (float): Factor for clipping.
  clip_weight (float): Weight for clipping iterations.
  norm_window (float): Normalization window in seconds.
  freqmin (float): Minimum frequency value.
  sampling_rate (float): Sampling rate.
  water_level (float): Water level for uniform filter.
  taper_length (int): Length of the cosine taper.

  Returns:
  array-like: Normalized signal.
  """

  # Convert the signal to float64
  sgn = sgn.astype(np.float64)

  if mode == 'clipping':
    # Clip the signal using a factor of standard deviation
    lim = clip_factor * np.std(sgn)
    sgn = np.clip(sgn, -lim, lim)

  elif mode == 'clipping_iteration':
    # Clip the signal iteratively using a factor of standard deviation
    lim = clip_factor * np.std(np.abs(sgn))
    while len(sgn[np.abs(sgn) > lim]) > 0:
      sgn[sgn > lim] /= clip_weight
      sgn[sgn < -lim] /= clip_weight

  elif mode == 'running_average':
    # Apply a running average normalization
    if norm_window is None:
      # Set norm_window to 0.5 / freqmin if not provided
      norm_window = 0.5 / freqmin

    # Calculate the half-window length and the window length
    hwin = int(sampling_rate * norm_window)
    lwin = 2 * hwin + 1

    # Apply uniform filter and add water_level to avoid divide by zero
    nmf = uniform_filter1d(np.abs(sgn), size=lwin,
                           mode='nearest') + water_level
    sgn /= nmf

    if taper_length is None:
      # Set taper_length to hwin if not provided
      taper_length = hwin

    # Apply cosine taper
    sgn = cosine_taper(sgn, taper_length)

  elif mode == '1bit':
    # Convert the signal to 1bit
    sgn = np.sign(sgn)

  elif mode == 'no_normalization':
    # Do nothing
    pass

  return sgn
```

```python
if __name__ == '__main__':

  # Define input parameters
  num_of_days = 10                                     # Number of days for which data will be processed
  input_data_dir = 'data'                              # Directory where input data is stored
  input_metadata_dir = 'metadata'                      # Directory where metadata is stored
  network = 'CI'                                       # Seismic network code
  stations = ['MLAC', 'PHL']                           # List of station codes to be processed
  channel = 'BHZ'                                      # Seismic channel to be processed (e.g., vertical component)
  year = 2004                                          # Year of the data to be processed
  date = UTCDateTime('2004-10-20T00:00:00.000000Z')    # Start date and time for the analysis

  # Define normalization parameters
  norm_mode = 'running_average'                        # Time-domain normalization mode (e.g., running_average, 1bit, clipping, etc.)
  response_output = 'VEL'                              # Output type after instrument correction (e.g., VEL for velocity and DISP for displacement)
  sampling_rate = 20.0                                 # Sampling rate in Hz
  freqmin = 0.001                                      # Minimum frequency for bandpass filtering
  freqmax = 0.9 * sampling_rate / 2.0                  # Maximum frequency for bandpass filtering (90% of the Nyquist frequency)
  max_lag_in_seconds = 2000.0                          # Maximum lag time in seconds for cross-correlation
  duration = 7200.0                                    # Duration of the signal window in seconds
  dspline = int(2 * sampling_rate / freqmin)           # Downsampling factor for spline interpolation
  taper_percentage = 0.001                             # Percentage of taper to apply at the edges of the signal
  tmin = -600.0                                        # Minimum time in the cross-correlogram (for plotting only)
  tmax =  600.0                                        # Maximum time in the cross-correlogram (for plotting only)
  tzoom_min = -200                                     # Minimum zoomed time in the cross-correlogram (for plotting only)
  tzoom_max = -40                                      # Maximum zoomed time in the cross-correlogram (for plotting only)

  # Calculate correlated signal parameters
  max_lag = int(sampling_rate * max_lag_in_seconds)    # Convert max lag from seconds to samples
  num_samples_in_day = int(sampling_rate * 86400 + 1)  # Number of samples in one day
  window_length = int(sampling_rate * duration)        # Convert signal window duration from seconds to samples
  npis = int(max_lag + 1)                              # Half of the number of points in the final stack
  delta = 1.0 / sampling_rate                          # Time interval between samples in seconds
  taper_length = int(0.5 * sampling_rate / freqmin)    # Length of the taper applied to the signal in samples
```

```python
  # Workflow 1
  start_time = time.time()

  # Print workflow information
  print ('Workflow 1')

  n1 = num_samples_in_day
  n2 = num_samples_in_day

  n = n2 - n1 + 1
  nfft = n1 + n2 - 1

  # Get response from station inventory
  response_1, freqs = inv1.get_response(f'{network}.{stations[0]}..{channel}',
                                        date).get_evalresp_response (t_samp=delta,
                                                                     nfft=nfft,
                                                                     output=response_output)
  response_2, freqs = inv2.get_response(f'{network}.{stations[1]}..{channel}',
                                        date).get_evalresp_response (t_samp=delta,
                                                                     nfft=nfft,
                                                                     output=response_output)

  # Calculate the necessary parameters
  n = 2 * max_lag + 1  # Number of samples in the cross-correlogram
  c = n // 2           # Center of the correlation window

  num_windows_in_a_day = num_samples_in_day // window_length  # Number of windows in a day

  shift = int(np.ceil(max_lag / window_length))  # Shift in samples for each window

  stack_WF1 = None                               # Stack for the data

  n1 = window_length                             # Number of samples in the first window
  n2 = window_length + 2 * max_lag               # Number of samples in the second window
  npts = n1 + n2 - 1                             # Total number of samples in the correlation window
  nrfft = sp.fft.next_fast_len(npts, real=True)  # Next fast length for the Fourier transform
  frange = freqmax - freqmin                     # Frequency range of interest

  # Print correlation information
  print(f'Correlating {stations[0]} and {stations[1]}...')

  # Read data
  st1_a = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_1.mseed')
  st2_a = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_1.mseed')

  # Apply spline interpolation
  st1_a[0].data = spline(st1_a[0].data, order=2, dspline=dspline)
  st2_a[0].data = spline(st2_a[0].data, order=2, dspline=dspline)

  # Check if the sampling rate of the data is the same as the desired sampling rate
  if int(st1_a[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
    # If not, resample the data
    st1_a.resample(sampling_rate)

  if int(st2_a[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
    st2_a.resample(sampling_rate)

  # Remove response from the data using the response object
  st1_a[0].data = removeResponse(st1_a[0].data, response_1, freqs,
                                 freqmin, freqmax, nfft, taper_length)
  st2_a[0].data = removeResponse(st2_a[0].data, response_2, freqs,
                                 freqmin, freqmax, nfft, taper_length)

  # Normalize the data
  st1_a[0].data = normalize(st1_a[0].data, norm_mode, freqmin, sampling_rate, taper_length)
  st2_a[0].data = normalize(st2_a[0].data, norm_mode, freqmin, sampling_rate, taper_length)

  # Repeat the same process for the other days of data
  st1_b = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_2.mseed')
  st2_b = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_2.mseed')

  st1_b[0].data = spline(st1_b[0].data, order=2, dspline=dspline)
  st2_b[0].data = spline(st2_b[0].data, order=2, dspline=dspline)

  if int(st1_b[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
    st1_b.resample(sampling_rate)

  if int(st2_b[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
    st2_b.resample(sampling_rate)

  st1_b[0].data = removeResponse(st1_b[0].data, response_1, freqs,
                                 freqmin, freqmax, nfft, taper_length)
  st2_b[0].data = removeResponse(st2_b[0].data, response_2, freqs,
                                 freqmin, freqmax, nfft, taper_length)

  st1_b[0].data = normalize(st1_b[0].data, norm_mode, freqmin, sampling_rate, taper_length)
  st2_b[0].data = normalize(st2_b[0].data, norm_mode, freqmin, sampling_rate, taper_length)

  st1_c = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_3.mseed')
  st2_c = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_3.mseed')

  st1_c[0].data = spline(st1_c[0].data, order=2, dspline=dspline)
  st2_c[0].data = spline(st2_c[0].data, order=2, dspline=dspline)

  if int(st1_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
    st1_c.resample(sampling_rate)

  if int(st2_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
    st2_c.resample(sampling_rate)

  st1_c[0].data = removeResponse(st1_c[0].data, response_1, freqs,
                                 freqmin, freqmax, nfft, taper_length)
  st2_c[0].data = removeResponse(st2_c[0].data, response_2, freqs,
                                 freqmin, freqmax, nfft, taper_length)

  st1_c[0].data = normalize(st1_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)
  st2_c[0].data = normalize(st2_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)

  # Initialize variables
  st1_response = st1_a + st1_b + st1_c
  st2_response = st2_a + st2_b + st2_c

  # Merge the traces
  st1_response.merge(method=0, fill_value='interpolate')
  st2_response.merge(method=0, fill_value='interpolate')

  # Get the response data
  sg1_response = st1_response[0].data
  sg2_response = st2_response[0].data

  # Loop through the days
  for i in range(num_of_days):

    print(f'Day {i + 1}')

    # If it is not the first or last day, update the data
    if i > 1 and i < num_of_days - 1:

      st1_a = st1_b
      st2_a = st2_b

      st1_b = st1_c
      st2_b = st2_c

      # Read the new data
      st1_c = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_{i + 2}.mseed')
      st2_c = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_{i + 2}.mseed')

      # Apply spline interpolation
      st1_c[0].data = spline(st1_c[0].data, order=2, dspline=dspline)
      st2_c[0].data = spline(st2_c[0].data, order=2, dspline=dspline)

      # Resample the data if necessary
      if int(st1_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
        st1_c.resample(sampling_rate)

      if int(st2_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
        st2_c.resample(sampling_rate)

      # Remove the response and normalize the data
      st1_c[0].data = removeResponse(st1_c[0].data, response_1, freqs,
                                     freqmin, freqmax, nfft, taper_length)
      st2_c[0].data = removeResponse(st2_c[0].data, response_2, freqs,
                                     freqmin, freqmax, nfft, taper_length)

      st1_c[0].data = normalize(st1_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)
      st2_c[0].data = normalize(st2_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)

      # Update the response data
      st1_response = st1_a + st1_b + st1_c
      st2_response = st2_a + st2_b + st2_c

      # Merge the traces
      st1_response.merge(method=0, fill_value='interpolate')
      st2_response.merge(method=0, fill_value='interpolate')

      # Get the updated response data
      sg1_response = st1_response[0].data
      sg2_response = st2_response[0].data

    if i == 0:

      # index for the first data segment
      index = 0

    elif i < num_of_days - 1:

      # index for the middle data segments
      index = 1 * (num_samples_in_day - 1)

    else:

      # index for the last data segment
      index = 2 * (num_samples_in_day - 1)

    # initialize first and last window
    first_window = 0
    last_window  = num_windows_in_a_day

    if i == 0:

      # if it's the first day, add shift to the first window
      first_window += shift

    elif i == num_of_days - 1:

      # if it's the last day, subtract shift from the last window
      last_window -= shift

    # loop through windows
    for j in range (first_window, last_window):

      # calculate indices for the current window
      i3 = index + j * (window_length - 1)
      i4 = i3 + window_length

      i1 = i3 - max_lag
      i2 = i4 + max_lag

      # get response data for the current window
      sg_1_response = sg1_response[i1:i2]
      sg_2_response = sg2_response[i3:i4]

      # calculate taper lengths
      taper_length1 = int (taper_percentage * n1)
      taper_length2 = int (taper_percentage * n2)

      # detrend the response data
      sg_1_response = sp.signal.detrend (sg_1_response)
      sg_2_response = sp.signal.detrend (sg_2_response)

      # apply the cosine taper to the response data
      sg_1_response = cosine_taper (sg_1_response, taper_length1)
      sg_2_response = cosine_taper (sg_2_response, taper_length2)

      # calculate lengths of the response data
      len1 = (sg_1_response.size - 1) * delta
      len2 = (sg_2_response.size - 1) * delta

      # calculate the Fourier transform of the response data
      sp1_response = sp.fft.rfft (sg_1_response[::-1], n = nrfft)
      sp2_response = sp.fft.rfft (sg_2_response, n = nrfft)

      # calculate the number of smooth points
      nsmo1 = int (np.fix (min (0.05, 0.5 * frange) * len1))
      nsmo2 = int (np.fix (min (0.05, 0.5 * frange) * len2))

      # calculate the frequency array
      f  = np.fft.rfftfreq (nrfft, delta)
      ii = ((f > freqmin) & (f < freqmax)).nonzero ()[0]

      # initialize the smoothed response data
      spw1_response = np.zeros (sp1_response.size, dtype = complex)

      # calculate the indices for the smoothed response data
      i1 = ii[0]
      i2 = ii[0] + nsmo1 + 1
      i3 = ii[-1] - nsmo1
      i4 = ii[-1] + 1

      # calculate the smoothed response data
      spw1_response[i2:i3] = np.ones (i3 - i2)
      spw1_response[i1:i2] = np.cos (np.linspace (np.pi / 2, np.pi, nsmo1 + 1)) ** 2
      spw1_response[i3:i4] = np.cos (np.linspace (0, np.pi / 2, nsmo1 + 1)) ** 2

      # adjust the phase of the smoothed response data
      spw1_response[i1:i4] *= np.exp (1j * np.angle (sp1_response[i1:i4]))

      # initialize the smoothed response data
      spw2_response = np.zeros (sp2_response.size, dtype = complex)

      # calculate the indices for the smoothed response data
      i1 = ii[0]
      i2 = ii[0] + nsmo2 + 1
      i3 = ii[-1] - nsmo2
      i4 = ii[-1] + 1

      # calculate the smoothed response data
      spw2_response[i2:i3] = np.ones (i3 - i2)
      spw2_response[i1:i2] = np.cos (np.linspace (np.pi / 2, np.pi, nsmo2 + 1)) ** 2
      spw2_response[i3:i4] = np.cos (np.linspace (0, np.pi / 2, nsmo2 + 1)) ** 2

      # adjust the phase of the smoothed response data
      spw2_response[i1:i4] *= np.exp (1j * np.angle (sp2_response[i1:i4]))

      # calculate the inverse Fourier transform of the smoothed response data
      sgw1_response = sp.fft.irfft (spw1_response, n = nrfft)
      sgw2_response = sp.fft.irfft (spw2_response, n = nrfft)

      # calculate the Fourier transform of the smoothed response data
      spw1_response = sp.fft.rfft (sgw1_response, n = nrfft)
      spw2_response = sp.fft.rfft (sgw2_response, n = nrfft)

      # calculate the convolution of the Fourier transforms of the response data
      conv_fft_response = spw1_response * spw2_response

      # calculate the inverse Fourier transform of the convolution
      array_WF1 = sp.fft.irfft (conv_fft_response, n = nrfft)[:npts]

      # center the array
      green_WF1 = centered (array_WF1, n2 - n1 + 1)

      # if it's the first window, initialize the stacked response data
      if stack_WF1 is None:

        stack_WF1 = green_WF1

      # else, add the current response data to the stacked response data
      else:

        stack_WF1 = np.add (stack_WF1, green_WF1)

  # Slice the array
  stack_WF1 = stack_WF1[c - npis + 1:c + npis]

  # Calculate the time array
  t = np.linspace (-max_lag_in_seconds, max_lag_in_seconds, 2 * npis - 1)

  # Write the stacked waveform to a text file
  print (f'Writing {network}.{stations[0]}_{stations[1]}..{channel}_stack_WF1.txt...')

  np.savetxt (f'{network}.{stations[0]}_{stations[1]}..{channel}_stack_WF1.txt',
              np.c_[t, stack_WF1], fmt = '%.6E')

  # Calculate the elapsed time for the workflow
  elapsed_time_WF1 = time.time () - start_time
```

```python
  # Workflow 2
  start_time = time.time()

  print('Workflow 2')

  # Calculate the necessary parameters
  n = 2 * max_lag + 1  # Number of samples in the cross-correlogram
  c = n // 2           # Center of the correlation window

  num_windows_in_a_day = num_samples_in_day // window_length  # Number of windows in a day

  shift = int(np.ceil(max_lag / window_length))  # Shift in samples for each window

  stack_fft_WF2 = None                           # Stack for the final Fourier transformed data

  n1 = window_length                             # Number of samples in the first window
  n2 = window_length + 2 * max_lag               # Number of samples in the second window
  npts = n1 + n2 - 1                             # Total number of samples in the correlation window
  nrfft = sp.fft.next_fast_len(npts, real=True)  # Next fast length for the Fourier transform
  frange = freqmax - freqmin                     # Frequency range of interest

  print(f'Correlating {stations[0]} and {stations[1]}...')

  # Read the data for the first three days
  st1_a = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_1.mseed')
  st2_a = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_1.mseed')

  # Perform spline interpolation and normalization on the data
  st1_a[0].data = spline(st1_a[0].data, order=2, dspline=dspline)
  st2_a[0].data = spline(st2_a[0].data, order=2, dspline=dspline)

  if int(st1_a[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
      st1_a.resample(sampling_rate)

  if int(st2_a[0].stats.sampling_rate) != int(sampling_rate * 1e3):
      st2_a.resample(sampling_rate)

  st1_a[0].data = normalize(st1_a[0].data, norm_mode, freqmin, sampling_rate, taper_length)
  st2_a[0].data = normalize(st2_a[0].data, norm_mode, freqmin, sampling_rate, taper_length)

  st1_b = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_2.mseed')
  st2_b = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_2.mseed')

  st1_b[0].data = spline(st1_b[0].data, order=2, dspline=dspline)
  st2_b[0].data = spline(st2_b[0].data, order=2, dspline=dspline)

  if int(st1_b[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
      st1_b.resample(sampling_rate)

  if int(st2_b[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
      st2_b.resample(sampling_rate)

  st1_b[0].data = normalize(st1_b[0].data, norm_mode, freqmin, sampling_rate, taper_length)
  st2_b[0].data = normalize(st2_b[0].data, norm_mode, freqmin, sampling_rate, taper_length)

  st1_c = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_3.mseed')
  st2_c = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_3.mseed')

  st1_c[0].data = spline(st1_c[0].data, order=2, dspline=dspline)
  st2_c[0].data = spline(st2_c[0].data, order=2, dspline=dspline)

  if int(st1_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
      st1_c.resample(sampling_rate)

  if int(st2_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
      st2_c.resample(sampling_rate)

  st1_c[0].data = normalize(st1_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)
  st2_c[0].data = normalize(st2_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)

  # Concatenate the data from the three days
  st1 = st1_a + st1_b + st1_c
  st2 = st2_a + st2_b + st2_c

  # Merge the data
  st1.merge(method=0, fill_value='interpolate')
  st2.merge(method=0, fill_value='interpolate')

  # Get the data as numpy arrays
  sg1 = st1[0].data
  sg2 = st2[0].data

  # Loop over the days
  for i in range(num_of_days):

    print(f'Day {i + 1}')

    # If it's not the first or last day, read the data from the next day
    if i > 1 and i < num_of_days - 1:

      # Update the data arrays
      st1_a = st1_b
      st2_a = st2_b
      st1_b = st1_c
      st2_b = st2_c

      # Read the data from the next day
      st1_c = read(f'{input_data_dir}/noise.{network}.{stations[0]}.{channel}.day_{i + 2}.mseed')
      st2_c = read(f'{input_data_dir}/noise.{network}.{stations[1]}.{channel}.day_{i + 2}.mseed')

      # Smooth the data
      st1_c[0].data = spline(st1_c[0].data, order=2, dspline=dspline)
      st2_c[0].data = spline(st2_c[0].data, order=2, dspline=dspline)

      # Resample the data if necessary
      if int(st1_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
        st1_c.resample(sampling_rate)

      if int(st2_c[0].stats.sampling_rate * 1e3) != int(sampling_rate * 1e3):
        st2_c.resample(sampling_rate)

      # Normalize the data
      st1_c[0].data = normalize(st1_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)
      st2_c[0].data = normalize(st2_c[0].data, norm_mode, freqmin, sampling_rate, taper_length)

      # Concatenate the data
      st1 = st1_a + st1_b + st1_c
      st2 = st2_a + st2_b + st2_c

      # Merge the data
      st1.merge(method=0, fill_value='interpolate')
      st2.merge(method=0, fill_value='interpolate')

      # Get the data as numpy arrays
      sg1 = st1[0].data
      sg2 = st2[0].data

    # If it's the first day, set the index to 0
    if i == 0:
      index = 0

    # If it's not the first or last day, set the index to the first window
    elif i < num_of_days - 1:
      index = 1 * (num_samples_in_day - 1)

    # If it's the last day, set the index to the second window
    else:
      index = 2 * (num_samples_in_day - 1)

    # Set the range of windows to process
    first_window = 0
    last_window = num_windows_in_a_day

    # If it's the first day, shift the first window
    if i == 0:
      first_window += shift

    # If it's the last day, shift the last window
    elif i == num_of_days - 1:
      last_window -= shift

    # Loop over the windows
    for j in range(first_window, last_window):

      # Calculate the indices for the current window
      i3 = index + j * (window_length - 1)
      i4 = i3 + window_length

      i1 = i3 - max_lag
      i2 = i4 + max_lag

      # Get the data for the current window
      sg_1 = sg1[i1:i2]
      sg_2 = sg2[i3:i4]

      # Taper the data
      taper_length1 = int(taper_percentage * n1)
      taper_length2 = int(taper_percentage * n2)

      sg_1 = sp.signal.detrend(sg_1)
      sg_2 = sp.signal.detrend(sg_2)

      sg_1 = cosine_taper(sg_1, taper_length1)
      sg_2 = cosine_taper(sg_2, taper_length2)

      # Calculate the lengths of the windows
      len1 = (sg_1.size - 1) * delta
      len2 = (sg_2.size - 1) * delta

      # Perform the Fourier transform
      sp1 = sp.fft.rfft(sg_1[::-1], n=nrfft)
      sp2 = sp.fft.rfft(sg_2, n=nrfft)

      # Apply the smoothing function
      nsmo1 = int(np.fix(min(0.05, 0.5 * frange) * len1))
      nsmo2 = int(np.fix(min(0.05, 0.5 * frange) * len2))

      f = np.fft.rfftfreq(nrfft, delta)
      ii = ((f > freqmin) & (f < freqmax)).nonzero()[0]

      spw1 = np.zeros(sp1.size, dtype=complex)

      i1 = ii[0]
      i2 = ii[0] + nsmo1 + 1
      i3 = ii[-1] - nsmo1
      i4 = ii[-1] + 1

      spw1[i2:i3] = np.ones(i3 - i2)
      spw1[i1:i2] = np.cos(np.linspace(np.pi / 2, np.pi, nsmo1 + 1)) ** 2
      spw1[i3:i4] = np.cos(np.linspace(0, np.pi / 2, nsmo1 + 1)) ** 2

      spw1[i1:i4] *= np.exp(1j * np.angle(sp1[i1:i4]))

      spw2 = np.zeros(sp2.size, dtype=complex)

      i1 = ii[0]
      i2 = ii[0] + nsmo2 + 1
      i3 = ii[-1] - nsmo2
      i4 = ii[-1] + 1

      spw2[i2:i3] = np.ones(i3 - i2)
      spw2[i1:i2] = np.cos(np.linspace(np.pi / 2, np.pi, nsmo2 + 1)) ** 2
      spw2[i3:i4] = np.cos(np.linspace(0, np.pi / 2, nsmo2 + 1)) ** 2

      spw2[i1:i4] *= np.exp(1j * np.angle(sp2[i1:i4]))

      # Calculate the convolution of the Fourier transforms
      conv_fft = spw1 * spw2

      # Update the stacked Fourier transform
      if stack_fft_WF2 is None:
        stack_fft_WF2 = conv_fft
      else:
        stack_fft_WF2 = np.add(stack_fft_WF2, conv_fft)

  # Calculate the response of each station
  response_a, freqs = inv1.get_response(f'{network}.{stations[0]}..{channel}',
                                        date).get_evalresp_response(t_samp = delta,
                                                                    nfft = nrfft,
                                                                    output = response_output)
  response_b, freqs = inv2.get_response(f'{network}.{stations[1]}..{channel}',
                                        date).get_evalresp_response (t_samp = delta,
                                                                     nfft = nrfft,
                                                                     output = response_output)

  # Calculate the frequencies of the prefilter
  f1 = freqmin
  f2 = 1.4 * freqmin
  f3 = 0.9 * freqmax
  f4 = freqmax

  # Create the prefilter
  prefilter = np.zeros_like(freqs)

  prefilter[(freqs >= f1) & (freqs <  f2)] = (freqs[(freqs >= f1) & (freqs < f2)] - f1) / (f2 - f1)
  prefilter[(freqs >= f2) & (freqs <= f3)] = 1.0
  prefilter[(freqs >  f3) & (freqs <= f4)] = (f4 - freqs[(freqs > f3) & (freqs <= f4)]) / (f4 - f3)

  # Apply the frequency response of each station to the stacked Fourier transform
  response = prefilter * response_a * response_b.conj()
  response_inverse_WF2 = np.exp(-1j * np.angle (response))

  stack_fft_WF2 = stack_fft_WF2 * response_inverse_WF2

  # Transform the stacked Fourier transform back to the time domain
  array_WF2 = sp.fft.irfft (stack_fft_WF2, n = nrfft)[:npts]

  # Apply the window function to the time domain signal
  stack_WF2 = centered(array_WF2, n2 - n1 + 1)
  stack_WF2 = stack_WF2[c - npis + 1:c + npis]

  # Calculate the time array
  t = np.linspace (-max_lag_in_seconds, max_lag_in_seconds, 2 * npis - 1)

  # Write the stacked waveform to a text file
  print(f'Writing {network}.{stations[0]}_{stations[1]}..{channel}_stack_WF2.txt...')

  np.savetxt(f'{network}.{stations[0]}_{stations[1]}..{channel}_stack_WF2.txt',
             np.c_[t, stack_WF2], fmt = '%.6E')

  # Calculate the elapsed time for the workflow
  elapsed_time_WF2 = time.time() - start_time
```
